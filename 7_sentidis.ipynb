#### 从个数算成percentage 也就是要除以推文数

```python
# 消极情绪
import pandas as pd
drug=['hcq','iver','zinc','vitd','rem']
#drug=['vitd']

for i in drug:
    n=pd.read_csv('freq'+i+"n.csv")
    s=pd.read_csv('freq'+i+"s.csv")
    neg=pd.merge(n, s, on='created_at')
    print(neg.head(1))
    #df_data['平均值'] =df_data.apply(lambda x: x['总数'] / x['人数'], axis=1)
    try:
        neg['per'] =neg.apply(lambda x: x['neg_num'] / x['senti_num'], axis=1)
        
    except ZeroDivisionError: # because some users has no text
        neg['per']=0
    neg.to_csv('freq'+i+"np.csv")
```

```python
# 积极情绪
import pandas as pd
drug=['hcq','iver','zinc','vitd','rem']
#senti=['pos','pos']
for i in drug:
    p=pd.read_csv('freq'+i+"p.csv")
    s=pd.read_csv('freq'+i+"s.csv")
    pos=pd.merge(p, s, on='created_at')
    print(pos.head(1))
    #df_data['平均值'] =df_data.apply(lambda x: x['总数'] / x['人数'], axis=1)
    try:
        pos['per'] =pos.apply(lambda x: x['pos_num'] / x['senti_num'], axis=1)   
    except ZeroDivisionError: # because some users has no text
        pos['per']=0
    pos.to_csv('freq'+i+"pp.csv",index=False)
```

```python
#随机抽出50条 手动评估 比较一致性
import pandas as pd
iver = pd.read_csv("iver.csv",lineterminator='\n',low_memory=False, keep_default_na=False)
iver20=iver.sample(n=20,random_state=1)
iver20.to_csv('iver20.csv')

rem = pd.read_csv("rem.csv",lineterminator='\n',low_memory=False, keep_default_na=False)
rem20=rem.sample(n=20,random_state=1)
rem20.to_csv('rem20.csv')

HCQ = pd.read_csv("HCQ.csv",lineterminator='\n',low_memory=False, keep_default_na=False)
HCQ20=HCQ.sample(n=20,random_state=1)
HCQ20.to_csv('HCQ20.csv')

vitd = pd.read_csv("vitd.csv",lineterminator='\n',low_memory=False, keep_default_na=False)
vitd20=vitd.sample(n=20,random_state=1)
vitd20.to_csv('vitd20.csv')

zinc = pd.read_csv("zinc.csv",lineterminator='\n',low_memory=False, keep_default_na=False)
zinc20=zinc.sample(n=20,random_state=1)
zinc20.to_csv('zinc20.csv')

df=pd.concat([iver20,HCQ20,vitd20,rem20,zinc20])
```
